## Transfer Learning - Basics
Making Predictions using Pre-trained Networks

```python
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
#create the model
model = ResNet50(weights='imagenet')
```

```python
# Read the Image
import cv2
img_path = 'image.jpg'
img = cv2.imread(img_path)
img = cv2.resize(img,(224,224))
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)#https://www.tutorialspoint.com/numpy/numpy_expand_dims.htm
x = preprocess_input(x)
```

```python
import matplotlib.pyplot as plt
plt.title("Actual Image")
plt.imshow(img)
plt.show()
plt.title("Normalised Image")
plt.imshow(x[0])
plt.show()
```

```python
preds = model.predict(x)
print('Predicted:', decode_predictions(preds, top=3)[0])#decode prediction will map ur prediction with the class labels
```
